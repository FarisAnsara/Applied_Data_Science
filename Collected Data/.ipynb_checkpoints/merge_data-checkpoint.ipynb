```python
import zipfile
import pandas as pd

zip_file_name = 'csv_data_2023.zip'
combined_data = pd.DataFrame()

with zipfile.ZipFile(zip_file_name, 'r') as zip_file:
    n = 1
    for file_name in zip_file.namelist():
        if file_name.endswith('.csv'):
            with zip_file.open(file_name) as csv_file:
                try:
                    df = pd.read_csv(csv_file, encoding='utf-8')
                    df.columns = df.columns.str.lower()

                    # Find common columns dynamically
                    common_columns = list(set(df.columns) & set(combined_data.columns))

                    # Check if 'occupancy' is in both DataFrames
                    if 'occupancy' in df.columns and 'occupancy' in combined_data.columns:
                        # Convert 'occupancy' column to string
                        df['occupancy'] = df['occupancy'].astype(str)

                    # Check if 'date' is in both DataFrames
                    if 'date' in df.columns and 'date' in combined_data.columns:
                        df['date'] = pd.to_datetime(df['date'], errors='coerce').dt.strftime('%d-%m-%Y')

                    # Convert 'Time' from 'HH:MM:SS' to 'hh.mm am/pm' format
                    if 'time' in df.columns:
                        df['time'] = pd.to_datetime(df['time'], errors='coerce').dt.strftime('%I.%M %p').str.lower().replace('pm', ' pm').replace('am', ' am')

                    # Convert 'Direction' values to lowercase, handling non-string values
                    if 'direction' in df.columns:
                        df['direction'] = df['direction'].astype(str).str.lower()                        

                    # Check if the DataFrame is empty or not
                    if combined_data.empty:
                        combined_data = df
                    else:
                        # Concatenate along rows using common columns
                        combined_data = pd.concat([combined_data, df], axis=0, ignore_index=True, join='outer', keys=common_columns)

                    n += 1
                except UnicodeDecodeError:
                    print(f"Error reading {file_name}: UnicodeDecodeError")

# Create the 'eco' column based on specified conditions
combined_data['eco'] = combined_data[eco_friendly_labels].apply(
    lambda row: any(str(val).lower() in ['yes', 'y', '1'] for val in row if pd.notna(val)) if any(pd.notna(val) for val in row) else None,
    axis=1
)
# Save the combined data to a CSV file
combined_data.to_csv('combined_data2.csv', index=False)
```

```python
import zipfile
import pandas as pd

zip_file_name = 'csv_data_2023.zip'
combined_data = pd.DataFrame()
eco_friendly_labels = ['is_env_friendly', 'eco-friendly', 'green badge', 'low emissions compliant', 'green vehicle', 'is_env_fr', 'ecofriendly']

with zipfile.ZipFile(zip_file_name, 'r') as zip_file:
    n = 1
    for file_name in zip_file.namelist():
        if file_name.endswith('.csv'):
            with zip_file.open(file_name) as csv_file:
                try:
                    df = pd.read_csv(csv_file, encoding='utf-8')
                    df.columns = df.columns.str.lower()

                    # Find common columns dynamically
                    common_columns = list(set(df.columns) & set(combined_data.columns))

                    # Check if 'occupancy' is in both DataFrames
                    if 'occupancy' in df.columns and 'occupancy' in combined_data.columns:
                        # Convert 'occupancy' column to string
                        df['occupancy'] = df['occupancy'].astype(str)

                    # Check if 'date' is in both DataFrames
                    if 'date' in df.columns and 'date' in combined_data.columns:
                        df['date'] = pd.to_datetime(df['date'], errors='coerce').dt.strftime('%d-%m-%Y')

                    # Convert 'Time' from 'HH:MM:SS' to 'hh.mm am/pm' format
                    if 'time' in df.columns:
                        df['time'] = pd.to_datetime(df['time'], errors='coerce').dt.strftime('%I.%M %p').str.lower().replace('pm', ' pm').replace('am', ' am')

                    # Convert 'Direction' values to lowercase, handling non-string values
                    if 'direction' in df.columns:
                        df['direction'] = df['direction'].astype(str).str.lower()

                    for col in df.columns:
                        if col in eco_friendly_labels:
                            # Check if 'eco' column exists, create it if not
                            if 'eco' not in combined_data.columns:
                                combined_data['eco'] = None  # Initialize the 'eco' column with None
                    
                            for val in df[col]:
                                if str(val).lower() in ['yes', 'y', '1']:
                                    combined_data['eco'] = True
                                else:
                                    combined_data['eco'] = False

                    # Check if the DataFrame is empty or not
                    if combined_data.empty:
                        combined_data = df
                    else:
                        # Concatenate along rows using common columns
                        combined_data = pd.concat([combined_data, df], axis=0, ignore_index=True, join='outer', keys=common_columns)

                    n += 1
                except UnicodeDecodeError:
                    print(f"Error reading {file_name}: UnicodeDecodeError")

# Save the combined data to a CSV file
combined_data.to_csv('combined_data2.csv', index=False)
```
